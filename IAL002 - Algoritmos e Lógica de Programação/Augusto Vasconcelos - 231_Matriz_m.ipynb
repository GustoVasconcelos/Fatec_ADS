### 1. Faça um programa que leia uma matriz 2x3 (2 linhas, 3 colunas). Apresente os elementos da matriz e seus respectivos índices.

```
m = []
for lin in range(2):
    linha = []
    for col in range(3):
        linha.append(int(input(f'Digite um número [{lin},{col}]: ')))
    m.append(linha)
for lin in range(len(m)): # len -> qtde de linhas
    print('\n+-----+-----+-----+')
    for col in range(len(m[0])): # len(matriz[0]) qtde de colunas
        print(f'|{m[lin][col]}|', end='\t') # \t = tabulação
print('\n+-----+-----+-----+')
```

### 2. Faça um programa que leia uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os salários/elementos que será o montante pago pela empresa a esses funcionários.

```
salario = []
soma = 0
for linha in range(2):
    linha = []
    for coluna in range(2):
        linha.append(float(input("Digite o salario do funcionário: ")))
    salario.append(linha)
print(f'\nTodos os salários: ')
for linha in range(len(salario)):
    for coluna in range(len(salario[0])):
        soma = soma + salario[linha][coluna]
        print(salario[linha][coluna], end="\t")
    print()
print(f'\nA soma dos salários será: {soma}')
```

### 3. Faça um programa para ler uma matriz 5x2 preços. Encontre e apresente os ÍNDICES dos valores menores que 23 reais.

```
precos = []
for linha in range(5):
    linha = []
    for coluna in range(2):
        linha.append(float(input("Digite o preço do produto: ")))
    precos.append(linha)
print('\nProdutos com preços menores que R$23.00')
for linha in range(len(precos)):
    for coluna in range(len(precos[0])):
        if precos[linha][coluna] < 23:
            print(f'[{linha},{coluna}]: R${precos[linha][coluna]:.2f}')
```

### 4. Faça um programa que leia números inteiros w e p e os elementos de uma matriz A de números inteiros de dimensão w x p e conte o número de elementos que são iguais a zero. Observação w representa a quantidade de linhas e p as colunas


```
matriz = []
zero = 0
w = int(input("Digite o número de linhas: "))
p = int(input("Digite o número de colunas: "))
print()
for lin in range(w):
    linha = []
    for col in range(p):
        linha.append(int(input("Digite um número: ")))
    matriz.append(linha)
for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if matriz[lin][col] == 0:
            zero += 1
if zero > 0:
    print(f'\nQuantidade de númros iguais a zero (0): {zero}')
else:
    print(f'\nSem números iguais a zero (0).')
```

### 5. Faça um programa que leia uma matriz 3 x 2, que representa preços de produtos, crie OUTRA matriz que armazene todos os preços com 7% de aumento.

```
precos, precos_aum = [], []
for lin in range(3):
    linha = []
    for col in range(2):
        linha.append(float(input(f'Digite o preço [{lin},{col}] R$: ')))
    precos.append(linha)
precos_aum = precos
print("\nApresentação da matriz preço.......")
for lin in range(len(precos)):
    for col in range(len(precos[0])):
        print(f'{precos[lin][col]}', end="\t")
    print()
print("\nApresentação da matriz aumento.......")
for lin in range(len(precos_aum)):
    for col in range(len(precos_aum[0])):
        precos_aum[lin][col] += precos_aum[lin][col] * 7 / 100
        print(f'{precos_aum[lin][col]}', end="\t")
    print()
```

###7. Faça um programa que leia uma matriz 4 x 4. 
###* Apresente os valores/elementos e seus respectivos índices (linhas e colunas) que estão na diagonal principal.
###* Calcule e apresente a soma dos elementos acima da diagonal principal, que estão em negrito e com a cor de fundo amarela.

image.png


```
matriz = []
soma = 0
for lin in range(4):
    linha = []
    for col in range(4):
        linha.append(int(input("Digite um valor: ")))
    matriz.append(linha)
print(46 * '-' + "\n| Os valores e indices da diagonal principal |\n" + 46 * '-')
for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if lin == col:
            print(f'Posição[{lin},{col}]: {matriz[lin][col]}')
        if lin < col:
            soma += matriz[lin][col]
print(59 * '-' + f"\n| A soma dos elementos acima da diagonal principal é: {soma} |\n" + 59 * '-')
```
