###1. Faça um algoritmo  que mostre 10 vezes a frase “Bem vindo a Fatec!”. Pode implementar com o comando while ou for.

```
print('Exemplo do while.............\n\n')
# 1º passo - inicialização
contador = 1
# 2º passo - condição/critério de parada
while contador <= 10:
    print(contador, 'Bem vindo a Fatec!')
    contador = contador + 1 # 3º passo - incremento
```

```
print('\n\nExemplo do for.............\n')
# 1º passo - inicialização, 2º passo - condição/critério de parada
# e 3º passo - incremento, numa única linha no for
for contador in range(1,11):
    print(contador,'Bem vindo a Fatec!')
```

###6. Faça um algoritmo que receba a idade de 10 pessoas, calcule e exiba a quantidade de pessoas maiores de idade, sendo que a maioridade é obtida após completar 18 anos. Pode implementar com o comando while ou for.


```
print('Com While\n\n')
cont = 1
maior_idade = 0
while cont <= 10:
    idade = int(input(f'Digite a idade da {cont}ª pessoa: '))
    if idade >= 18:
        maior_idade += 1
    print(idade)
    cont += 1
print(f'Pessoas maiores de idade: {maior_idade}')
```

```
print('Com For\n\n')
maior_idade = 0
for cont in range(1,11):
    idade = int(input(f'Digite a idade da {cont}ª pessoa: '))
    if idade >= 18:
        maior_idade += 1
    print(idade)
print(f'Pessoas maiores de idade: {maior_idade}')
```

###7. Escreva um algoritmo que receba 23 números, calcule e exiba a quantidade de números pares e impares. Pode implementar com o comando while ou for.


```
print('Com While\n\n')
cont = 1
par = 0
impar = 0
while cont <= 23:
    numero = int(input('Digite um número: '))
    if numero % 2 == 0:
        par += 1        
    else:
        impar += 1
    cont += 1
print(f'Quantidade de numeros pares: {par}\nQuantidade de numeros impares: {impar}')
```

```
print('Com For\n\n')
par = 0
impar = 0
for cont in range(0,23):
    numero = int(input('Digite um número: '))
    if numero % 2 == 0:
        par += 1        
    else:
        impar += 1
print(f'Quantidade de numeros pares: {par}\nQuantidade de numeros impares: {impar}')
```

###8. Faça um algoritmo que calcule e exiba o salário reajustado de dez funcionários de acordo com a seguinte regra (pode implementar com o comando while ou for): 
## Salário até 300, reajuste de 50%; 

##Salários maiores que 300, reajuste de 30%.




```
print('Com While\n\n')
cont = 0
while cont < 10:
    salario = float(input('Digite o salário do funcionário: '))
    if salario > 0 and salario <= 300:
        reajuste = salario + (salario * 50 / 100)
        cont += 1
        print(f'O salário do funcionário reajustado ficou em R${reajuste:.2f}')
    elif salario > 300:
        reajuste = salario + (salario * 30 / 100)
        print(f'O salário do funcionário reajustado ficou em R${reajuste:.2f}')
        cont += 1
    else:
        print('Digite um valor maior que zero!')
```

```
print('Com For\n\n')
for cont in range(10):
    salario = float(input('Digite o salário do funcionário: '))
    if salario > 0 and salario <= 300:
        reajuste = salario + (salario * 50 / 100)
        print(f'O salário do funcionário reajustado ficou em R${reajuste:.2f}')
    elif salario > 300:
        reajuste = salario + (salario * 30 / 100)
        print(f'O salário do funcionário reajustado ficou em R${reajuste:.2f}')
    else:
        print('Digite um valor maior que zero!')
```

###12. Faça um programa que receba a idade e a altura de 20 pessoas. Calcule e exiba a média das alturas das pessoas com mais de 20 anos.




```
print('Com While\n\n')
contador = 1
soma_alturas = 0
soma_pessoas = 0
while contador <= 20:
    idade = int(input(f'Digite a idade da {contador}ª pessoa: '))
    altura = float(input(f'Digite a altura da {contador}ª pessoa: '))
    if idade > 20:
        soma_alturas = soma_alturas + altura
        soma_pessoas += 1
    contador += 1
media_alturas = soma_alturas / soma_pessoas
print(f'A média das alturas das pessoas com mais de 20 anos é {media_alturas:.2f}')
```

```
print('Com For\n\n')
soma_alturas = 0
soma_pessoas = 0
for contador in range(1,21):
    idade = int(input(f'Digite a idade da {contador}ª pessoa: '))
    altura = float(input(f'Digite a altura da {contador}ª pessoa: '))
    if idade > 20:
        soma_alturas = soma_alturas + altura
        soma_pessoas += 1
media_alturas = soma_alturas / soma_pessoas
print(f'A média das alturas das pessoas com mais de 20 anos é {media_alturas:.2f}')
```

###13. Faça um algoritmo que leia o preço de 20 TV, determine e apresente a média dos preços que possuem valor maior que R$ 1000. Pode implementar com o comando while ou for.

```
print('Com While\n\n')
contador = 1
conta_tv = 0
soma_preco_tv = 0
while contador <= 20:
    preco_tv = float(input(f'Digite o valor da {contador}ª TV: '))
    if preco_tv > 1000:
        conta_tv += 1
        soma_preco_tv = soma_preco_tv + preco_tv
    contador += 1
media_preco_tv = soma_preco_tv / conta_tv
print(f'A média do preço das TVs que custam mais de R$1000,00 é R${media_preco_tv:.2f}')
```

```
print('Com For\n\n')
conta_tv = 0
soma_preco_tv = 0
for contador in range(1,6):
    preco_tv = float(input(f'Digite o valor da {contador}ª TV: '))
    if preco_tv > 1000:
        conta_tv += 1
        soma_preco_tv = soma_preco_tv + preco_tv
media_preco_tv = soma_preco_tv / conta_tv
print(f'A média do preço das TVs que custam mais de R$1000,00 é R${media_preco_tv:.2f}')
```

###14. Faça um programa que receba a altura e nome  de 5 pessoas. Encontre e apresente a altura da pessoa mais alta e da mais baixa e seus respectivos nomes.

```
print('Com While\n\n')
contador = 1
while contador <= 5:
    nome = input(f'Digite o nome da {contador}ª pessoa: ')
    altura = float(input(f'Digite a altura da {contador}ª pessoa: '))
    if contador == 1:
        alto = altura
        nome_alto = nome
        baixo = altura
        nome_baixo = nome
    if altura > alto:
        alto = altura
        nome_alto = nome
    if altura < baixo:
        baixo = altura
        nome_baixo = nome
    contador += 1
print(f'Pessoa mais alta - {nome_alto} com {alto:.2f} de altura.\nPessoa mais baixa - {nome_baixo} com {baixo:.2f} de altura.')
```

###17. Faça um programa que calcule e apresente as tabuadas do 3 ao 7. Este exercício utiliza DUAS estruturas de repetição.

```
print('Com While\n\n')
x=3
while x <= 7:
    print(f'Tabuada do {x}')
    y = 1
    while y <= 10:
        print(f'{x} x {y} = {x * y}')
        y += 1
    x +=1
    print()
```

```
print('Com For\n\n')
for x in range(3,8):
    print(f'Tabuada do {x}')
    for y in range(1,11):
        print(f'{x} x {y} = {x * y}')
    print()
```

###18. Faça um programa para o curso de ADS (6 módulos), cada sala tem 30 alunos, calcule e apresente os seguintes itens:
*  Quantidade de homens e mulheres de cada módulo;
*  Média de idades de cada módulo;
*  Quantidade de homens e mulheres do curso todo;
*  Média de idades do curso todo.

Observação: este exercício utiliza DUAS estruturas de repetição.

```
print('Com For\n\n')
gen_fem_curso = 0
gen_masc_curso = 0
idades_curso = 0
qtde_alunos_curso = 0
for modulo in range(1,7):
    print(modulo,'º módulo','.'*50)
    # contar genero feminino = 0
    # inicializar as variaveis para este for
    gen_fem = 0
    gen_masc = 0
    idades_modulo = 0
    for alunos in range(1,3):
        # digitar os dados dos alunos       
        genero = input('Cadastre o gênero M/F: ')
        if genero == 'F' or genero == 'f':
            gen_fem += 1
        if genero == 'M' or genero == 'm':
            gen_masc += 1
        idade = int(input('Digite a idade: '))
        idades_modulo = idades_modulo + idade
    # mostrar resultados de cada módulo
    media_idades_modulo = idades_modulo / alunos
    print(f'Quantidade de Homens no {modulo}º modulo: {gen_masc}\nQuantidade de Mulheres no {modulo}º modulo: {gen_fem}')
    print(f'A média de idade do {modulo}º modulo é {media_idades_modulo:.2f}')
    # acumular valores para o curso e resetar variaveis
    idades_curso = idades_curso + idades_modulo
    gen_fem_curso = gen_fem_curso + gen_fem
    gen_masc_curso = gen_masc_curso + gen_masc
    qtde_alunos_curso = qtde_alunos_curso + alunos
# mostrar resultados do curso
media_idades_curso = idades_curso / qtde_alunos_curso
print('.'*50,'\nEstatísticas do Curso')
print('.'*50,)
print(f'Quantidade de Mulheres no curso todo: {gen_fem_curso}\nQuantidade de Homens no curso todo: {gen_masc_curso}')
print('-'*50,)
print(f'A média de idade de todos os alunos do curso é: {media_idades_curso:.2f}')
```

```
print('Com While\n\n')
gen_fem_curso = 0
gen_masc_curso = 0
idades_curso = 0
qtde_alunos_curso = 0
modulo = 1
while modulo <= 6:
    print(modulo,'º módulo','.'*50)
    # contar genero feminino = 0
    # inicializar as variaveis para este for
    gen_fem = 0
    gen_masc = 0
    idades_modulo = 0
    alunos = 1
    while alunos <= 2:
        # digitar os dados dos alunos       
        genero = input('Cadastre o gênero M/F: ')
        if genero == 'F' or genero == 'f':
            gen_fem += 1
        if genero == 'M' or genero == 'm':
            gen_masc += 1
        idade = int(input('Digite a idade: '))
        idades_modulo = idades_modulo + idade
        alunos  = alunos + 1
    # mostrar resultados de cada módulo
    media_idades_modulo = idades_modulo / (gen_fem + gen_masc)
    print(f'Quantidade de Homens no {modulo}º modulo: {gen_masc}\nQuantidade de Mulheres no {modulo}º modulo: {gen_fem}')
    print(f'A média de idade do {modulo}º modulo é {media_idades_modulo:.2f}')
    # acumular valores para o curso e resetar variaveis
    idades_curso = idades_curso + idades_modulo
    gen_fem_curso = gen_fem_curso + gen_fem
    gen_masc_curso = gen_masc_curso + gen_masc
    qtde_alunos_curso = gen_fem_curso + gen_masc_curso
    modulo = modulo + 1
# mostrar resultados do curso
media_idades_curso = idades_curso / qtde_alunos_curso
print('.'*50,'\nEstatísticas do Curso')
print('.'*50,)
print(f'Quantidade de Mulheres no curso todo: {gen_fem_curso}\nQuantidade de Homens no curso todo: {gen_masc_curso}')
print('-'*50,)
print(f'A média de idade de todos os alunos do curso é: {media_idades_curso:.2f}')
```
