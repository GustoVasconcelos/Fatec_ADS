### 1. Faça um programa que calcule e apresente a média de idades de uma sala de 35 alunos.​​

```
i, soma_idades = 0, 0
idade = []
while i < 5:
    idade.append(int(input('Digite a idade: ')))
    soma_idades = soma_idades + idade[i]
    i = i + 1
if i > 0:
    media = soma_idades / i
    print(f'A média de idades é {media:.2f} anos.')
```

### 9. Criar um algoritmo que leia dados para um vetor de 100 elementos inteiros, imprimir o maior, o menor, o percentual de números pares e a média dos elementos do vetor.
```
vet_int = []
qtde_pares, soma_elementos = 0, 0
for i in range(10):
    vet_int.append(int(input('Digite um número inteiro: ')))
    # soma dos elementos para tirar a média
    soma_elementos = soma_elementos + vet_int[i]
    # verifica quantidade de numeros pares
    if vet_int[i] % 2 == 0:
        qtde_pares += 1
    # verifica qual o maior e menor número
    if i == 0:
        num_maior = vet_int[i]
        num_menor = vet_int[i]
    if vet_int[i] > num_maior:
        num_maior = vet_int[i]
    if vet_int[i] < num_menor:
        num_menor = vet_int[i]
percentual = qtde_pares * 100 / len(vet_int)
media = soma_elementos / len(vet_int)
print(f'Maior número: {num_maior}\nMenor número: {num_menor}\nPercentual de números pares: {percentual:.2f}%\nMedia dos elementos do vetor: {media:.2f}')
```

### 23. Um vetor é palíndromo se ele não se alterar quando o mesmo for invertido. Escreva um programa que verifique se um vetor é palíndromo, fazendo comparação de índice/posição por índice/posição do vetor original (vo) com o vetor invertido (vi). Não use nenhuma função pronta da linguagem Python. A tamanho do vetor pode ser da sua escolha.
Exemplo:
vetor original    vo = {1, 3, 5, 2, 2, 5, 3, 1}  
vetor invertido é vi = {1, 3, 5, 2, 2, 5, 3, 1}
O vetor invertido **é palíndromo**, pois ele invertido é igual ao original

vetor original    vo = {9, 7, 5, 2, 4, 5, 3, 6}  
vetor invertido é vi = {6, 3, 5, 4, 2, 5, 7, 9}
O vetor invertido **não é palíndromo**, pois ele invertido é igual ao original



```
#percorrendo indice por indice e comparando eles
vo, vi = [], []
verdade = 0
for i in range(5):
    vo.append(int(input("Digite um número: ")))
cont = len(vo) - 1
for i in range(len(vo)):
    vi.append(vo[cont])
    cont -= 1
for i in range(len(vo)):
    if vo[i] == vi[i]:
        verdade += 1
if verdade == len(vo):
    print("O vetor é um palíndromo!")
else:
    print("O vetor NÃO é um palíndromo!")
```
