### 1. Faça um programa que leia um número, multiplique ele por 2 e apresente o resultado.


```
# 1º passo - identificar a(s) entrada(s) - variável número
numero = float(input("Digite um número: "))
# 2º passo - identificar o processamento/cálculo
valor = numero * 2
# 3º passo - apresentar a(s) variáveis que foram calculadas
print(f"O resultado é: {valor:.2f}")
```

###2. Faça um programa para calcular e apresentar a média de um aluno de ADS, referente a duas provas que foram realizadas. Mostre uma frase se ele foi aprovado ou reprovado.

```
# Digite seu programa aqui.
nota1 = float(input("Digite o valor da nota da primeira prova: "))
nota2 = float(input("Digite o valor da nota da segunda prova: "))
media = (nota1 + nota2) / 2
if media >= 6:
    print(f"Aluno aprovado, média: {media:.2f}")
else:
    print(f"Aluno reprovado, média: {media:.2f}")
```

### 3. Faça um programa que calcule e apresente a tabuada do número 7.


```
# Digite seu programa aqui.
# com contador while
cont, tab = 1, 7
while cont <= 10:
    print(f"{tab} x {cont} : {tab * cont}")
    cont += 1
```

```
# com contador for
for tabuada in range(1,11):
    print(f"7 x {tabuada} : {7 * tabuada}")
```

### 4. Faça um programa que armazene em um vetor 8 números. Calcule e mostre quantos elementos são positivos.

```
# Digite seu programa aqui.
vetor, positivos = [], 0
for i in range(8):
    vetor.append(float(input("Digite o número: ")))
print("Números positivos: ", end=" ")
for i in range(8):
    if vetor[i] > 0:
        positivos += 1
        print(f"|{vetor[i]}|", end=" ")
print()
print(f"{positivos} números são positivos")
```

```
vetor, positivos = [], 0
for i in range(8):
    vetor.append(float(input("Digite o número: ")))
for i in range(len(vetor)):
    if vetor[i] > 0:
        positivos += 1
print(f"{positivos} números são positivos")
```

```
vetor, positivos = [], 0
for i in range(8):
    vetor.append(float(input("Digite o número: ")))
for i in range(len(vetor)):
    if vetor[i] > 0:
        positivos += 1
print(f"{positivos} números são positivos")
```

### 5. Faça um programa para armazenar uma matriz 3x3 de elementos inteiros e conte e apresente quantos elementos são iguais a 3. Ao final, apresente toda a matriz.

```
matriz = []
for linha in range(3):
    linha_coluna = []
    for coluna in range(3):
        linha_coluna.append(int(input("Digite um número inteiro: ")))
    matriz.append(linha_coluna)
cont = 0
for linha in range(len(matriz)): #quantidades de linhas
    for coluna in range(len(matriz[0])): # quantidades de colunas
        print(matriz[linha][coluna], end='\t')
        if matriz[linha][coluna] == 3:
            cont += 1
    print()
print(f'Foram encontrados {cont} elementos iguais a 3.')
```
